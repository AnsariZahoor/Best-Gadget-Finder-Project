### Preparing the dataset before deployment

```python
import pandas as pd
import csv
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read data
df = pd.read_csv('MappedData.csv')
df.head(2)
```

```python
# Fill Not Available if value is empty
df = df.fillna('Not Available')
```

```python
# drop unwanted columns
df = df.drop(['Amazon_Name','Flipkart_Name','Snapdeal_Name','Amazon_OriginalPrice','Flipkart_OriginalPrice','Snapdeal_OriginalPrice','Amazon_Rating','Flipkart_Rating','Snapdeal_Rating'], axis = 1)
```

```python
df['Amazon_SalesPrice'] = df['Amazon_SalesPrice'].str.replace('Unavailable','Not Available')
```

```python
df['Flipkart_SalesPrice'] = df['Flipkart_SalesPrice'].str.replace('.0','')
```

```python
df.head()
```

```python
# save dataframe into csv
df.to_csv('FinalProductDataset.csv',index=0)
```
