### 1. Create new column as availablity
### 2. Create a unique ID, Snapdeal_1, Snapdeal_2.....
### 3. Generatae multiple names


```python
# importing libraries
import numpy as np
import pandas as pd 
import re
import csv
```

```python
dataset = pd.read_csv('SnapDealProducts.csv')
dataset.head()
```

## Add Availability

```python
def fuc1(r):
    """Return availability"""
    return 'Available'

def fuc2(r):
    """Return availability"""
    return 'Sold out'

dataset['Availability'] = dataset.apply(lambda row: fuc1(row) if row.notnull().all() else fuc2(row), axis=1)
```

## Add Unique ID

```python
Snapdeal_id = []
for i in range(1,len(dataset)+1):
    Id = f'Snapdeal_{i}'
    Snapdeal_id.append(Id)

dataset.insert(0, "ID",Snapdeal_id, True)
```

## Create Synonyms

```python
first_synonym = []
second_synonym = []
third_synonym = []

# iterate product name
for mystring in dataset['Name']:

    first_partitioned_string = mystring.rpartition('(')
    first_name = first_partitioned_string[0]
    first_synonym.append(first_name.strip())

    second_partitioned_string = mystring.rpartition(') ')
    second_name = second_partitioned_string[0]
    second_synonym.append(second_name.strip()+')')

    third_name = re.sub("[(].*[) ]", "", mystring)
    third_synonym.append(third_name.strip())
```

```python
# Create dataframe 
df = pd.DataFrame({'first_synonym': first_synonym,
                   'second_synonym': second_synonym,
                  'third_synonym': third_synonym})


# Merage all dataframe
df['Synonym'] = df[['first_synonym', 'second_synonym','third_synonym']].apply(lambda x: ' | '.join(x), axis=1)

# create a single list
synonym = df['Synonym'].to_list()
```

```python
# insert synonym list in dataframe
dataset.insert(2, "Synonyms",synonym, True)
dataset
```

```python
# save dataframe as csv file
dataset.to_csv('SnapdealDataset.csv',index=0)
```
